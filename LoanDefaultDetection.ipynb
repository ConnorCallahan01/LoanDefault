```python
import pandas as pd

attempt_to_read = pd.read_csv('2020_lar.csv', sep='|', encoding='ISO-8859-1')
```

```python
cut_df = attempt_to_read[1:200000]
```

```python
needed_df = pd.DataFrame(columns=['derived_ethnicity', 'derived_race', 'derived_sex', 'loan_type', 'loan_amount', 'interest_rate', 'total_loan_costs', 'lender_credits', 'loan_term', 
                                    'property_value', 'income', 'debt_to_income_ratio', 'applicant_credit_score_type'])
needed_df['derived_ethnicity'] = cut_df['derived_ethnicity']
needed_df['derived_race'] = cut_df['derived_race']
needed_df['derived_sex'] = cut_df['derived_sex']
needed_df['loan_type'] = cut_df['loan_type']
needed_df['loan_amount'] = cut_df['loan_amount']
needed_df['interest_rate'] = cut_df['interest_rate']
needed_df['total_loan_costs'] = cut_df['total_loan_costs']
needed_df['lender_credits'] = cut_df['lender_credits']
needed_df['loan_term'] = cut_df['loan_term']
needed_df['property_value'] = cut_df['property_value']
needed_df['income'] = cut_df['income']
needed_df['debt_to_income_ratio'] = cut_df['debt_to_income_ratio']
needed_df['applicant_credit_score_type'] = cut_df['applicant_credit_score_type']
```

```python
needed_df
```

```python
# Create classes for ethicity, race, and sex
ethnic_dict = {}
counter = 1
for i in needed_df['derived_ethnicity']:
    if i not in ethnic_dict.keys():
        ethnic_dict[i] = counter
        counter += 1

race_dict = {}
counter = 1
for i in needed_df['derived_race']:
    if i not in race_dict.keys():
        race_dict[i] = counter
        counter += 1

sex_dict = {}
counter = 1
for i in needed_df['derived_sex']:
    if i not in sex_dict.keys():
        sex_dict[i] = counter
        counter += 1

```

```python
for i in needed_df.index:
    if needed_df.loc[i]['derived_ethnicity'] in ethnic_dict.keys():
        needed_df.loc[i]['ethnicity'] = ethnic_dict[needed_df.loc[i]['derived_ethnicity']]
```

```python
for i in needed_df['derived_race']:
    if i in race_dict.keys():
        needed_df['race'] = race_dict[i]
```

```python
for i in needed_df['derived_sex']:
    if i in sex_dict.keys():
        needed_df['sex'] = sex_dict[i]
```

```python
needed_df = needed_df.replace(needed_df['derived_ethnicity'], needed_df['ethnicity'])
needed_df = needed_df.replace(needed_df['derived_race'], needed_df['race'])
needed_df = needed_df.replace(needed_df['derived_sex'], needed_df['sex'])

needed_df
```

```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

loanCorr = needed_df.corr()
#visualize the data to see where the correlations exist
plt.subplots(figsize=(20,20))
sns.heatmap(
    loanCorr,
    vmin=-1, vmax=1, center=0,
    cmap=sns.diverging_palette(22, 220, n=150),
    annot=True, annot_kws={'size': 8},
    square=True
)
```
